```python
import matplotlib.pyplot as plt
import numpy as np
import math

def gibbs_mix(n,T,x):
  R=8.314 # J / mol-K
  xA=x
  xB=1-x
  return n*R*T*(xA*math.log(xA)+xB*math.log(xB))

T=273.15 # Kelvin
n=1 # moles
x = np.arange(0.01, 1, 0.01)

y = []
for i in x:
  y.append(gibbs_mix(n,T,i))

plt.plot(x, y)
plt.xlabel(r'$x_A$')
plt.ylabel(r'$\Delta_{mix}G$')
plt.title('Gibbs free energy of mixing')
plt.show()
```
