```python
from IPython import get_ipython
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.widgets import Slider

def gibbs_mix(n,T,x,excess):
  R=8.314 # J / mol-K
  xA=x
  xB=1-x
  return n*R*T*(xA*math.log(xA)+xB*math.log(xB)+excess*xA*xB)

# Initial values
n=1 # moles
T_init = 273.15 # Kelvin
excess_init = 0 # Deviation from ideal solution behavior
x = np.arange(0.01, 1, 0.01) # Mol fraction of one component
y = []
for i in x:
  y.append(gibbs_mix(n,T_init,i,excess_init)/1000) # Initial Gibbs energy

# Create the figure and the line that we will manipulate
fig, ax = plt.subplots(num='Gibbs Free Energy of Mixing')
line, = ax.plot(x, y, lw=1) # Initial plot
ax.set_xlim(0, 1)
ax.set_ylim(-2, 2)
ax.set_xlabel(r'$x_A$')
ax.set_ylabel(r'$\Delta_{mix}G$ (kJ/mol)')
ax.set_title('Molar Gibbs free energy of mixing')

# adjust the main plot to make room for the sliders
fig.subplots_adjust(bottom=0.25)

# Make a horizontal slider to control the temperature.
axtemp = fig.add_axes([0.25, 0.1, 0.65, 0.03])
axexcess = fig.add_axes([0.25, 0.05, 0.65, 0.03])
temp_slider = Slider(
    ax=axtemp,
    label='Temperature (K)',
    valmin=200,
    valmax=400,
    valinit=T_init,
)
excess_slider = Slider(
    ax=axexcess,
    label=r'$\xi$',
    valmin=0,
    valmax=4,
    valinit=excess_init,
)
# The function to be called anytime a slider's value changes
def update(val):
    y_update = []
    for i in x:
         y_update.append(gibbs_mix(n,temp_slider.val,i,excess_slider.val)/1000)
    line.set_ydata(y_update)
    fig.canvas.draw_idle()

# register the update function with each slider
temp_slider.on_changed(update)
excess_slider.on_changed(update)

plt.show()
```
