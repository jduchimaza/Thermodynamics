```python
import numpy as np
import matplotlib.pyplot as plt

R=0.08206 # gas constant in (atm*L)/(mol*K)

# Calculate the pressure using the ideal gas law
def ideal_pressure(Vm,T):
  p = R*T/Vm
  return p

# Calculate the pressure using the van der Waals equation
def vdw_pressure(Vm,T):
  a=3.610   # Vdw constant for CO2 in atm*L^2/mol^2
  b=0.0429  # Vdw constant for CO2 in L/mol
  p = R*T/(Vm-b) - a/(Vm**2)
  return p

inputs = np.arange(0.05, 0.6, 0.01)
T = 293.15 # set the temperature to 0 K
T += 10    # specify the temperature in Â°C

outputs_ideal = []
outputs_vdw = []
for Vm in inputs:
  outputs_ideal.append(ideal_pressure(Vm,T))
  outputs_vdw.append(vdw_pressure(Vm,T))

# Plot the molar volumes vs pressure and format the graph
plt.plot(inputs,outputs_ideal)
plt.plot(inputs,outputs_vdw, color='#7F131A')
plt.xlabel('$V_m$ (L/mol)')
plt.ylabel('Pressure (atm)')
plt.grid()
plt.legend(['Ideal','Vdw'])
ax = plt.gca()
ax.set_ylim([10, 140])
ax.set_xlim([0.05, 0.6])
plt.show()

```
